###### Department wise codekata performence (pie chart)

```python
dept_codekata = college[['Department','CodeKata Score']]
dept_codekata.groupby(['Department']).sum().plot(kind='pie', y='CodeKata Score',autopct='%1.0f%%',figsize=(20,10))
```

###### Department wise toppers (horizantal bar graph or any visual representations of your choice)

```python
CS = college[college["Department"]=="Computer Science and Engineering"]
CS.sort_values(["CodeKata Score"],ascending = False,inplace=True)
X = CS["Name"].head(5)
Y = CS["CodeKata Score"].head(5)
plt.figure(figsize=(8,5), dpi=70)
plt.bar(X,Y)
plt.title("Computer Science and Engineering")
plt.show()
```
